```python
import numpy as np
import matplotlib.pyplot as plt

def gauss(x, m, sigma):
    #Funkcja Gaussa
    return np.exp(-((x - m) ** 2) / (2 * sigma ** 2))

# Dziedzina
x = np.linspace(2, 8, 1000)

# Parametry Gaussa
m, sigma = 5, 0.35

# Oblicz wartości (m - środek, sigma - "szerokość" opisu)
y = gauss(x, m, sigma)

# Wykres
plt.plot(x, y, label=f"Gauss(m={m}, σ={sigma})")
plt.xlabel("x")
plt.ylabel("μ(x)")
plt.title("Gaussowska funkcja przynależności")
plt.grid(True)
plt.legend()
plt.show()
```

```python
a,b,c,d = 5,3.3,1.4,0.2

p1 = gauss(a, 5, 0.35)
p2 = gauss(b, 3.4, 0.375)
p3 = gauss(c, 1.46, 0.17)
p4 = gauss(d, 0.2, 0.1)

p5 = gauss(a, 5.9, 0.51)
p6 = gauss(b, 2.8, 0.31)
p7 = gauss(c, 4.35, 0.465)
p8 = gauss(d, 1.3, 0.19)

p9 = gauss(a, 6.5, 0.63)
p10 = gauss(b, 3, 0.32)
p11 = gauss(c, 5.55, 0.55)
p12 = gauss(d, 2, 0.27)

psetosa = p1 + p2 + p3 + p4
pversi = p5 + p6 + p7 + p8
pvirg = p9 + p10 + p11 + p12


print('Podobieństwo do określenia "setosa":', psetosa)
print('Podobieństwo do określenia "versicolor":', pversi)
print('Podobieństwo do określenia "virginica":', pvirg)
```
